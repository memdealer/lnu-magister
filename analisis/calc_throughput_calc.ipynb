```python
import matplotlib.pyplot as plt
import pandas as pd

# Завантаження даних з CSV
file_path = 'data/throughput_unfiltered_data.csv'  # Вкажіть шлях до вашого файлу
data = pd.read_csv(file_path)

# Зчитування даних
requests = data['Requests']
rest_throughput = data['REST_Throughput']
grpc_throughput = data['gRPC_Throughput']

# Побудова графіка
plt.figure(figsize=(12, 7))
plt.plot(requests, rest_throughput, label='REST API Throughput', color='red')
plt.plot(requests, grpc_throughput, label='gRPC Throughput', color='green')
plt.title('Пропускна спроможність в залежності від кількості одночасних запитів')
plt.xlabel('Кількість одночасних запитів')
plt.ylabel('Пропускна спроможність (запити за секунду)')
plt.legend()
plt.grid(True)
plt.show()
```
