```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Завантаження даних з CSV
file_path = 'data/erlang_test_data.csv'  # Вкажіть шлях до вашого файлу
data = pd.read_csv(file_path)

# Зчитування даних
load = data['Load']
b_erlang_rest = data['Erlang_B_REST']
b_erlang_grpc = data['Erlang_B_gRPC']

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.plot(load, b_erlang_rest, label='REST API (каналів: 10)', color='red')
plt.plot(load, b_erlang_grpc, label='gRPC (каналів: 50)', color='green')
plt.title('Ймовірність втрати запитів за формулою Ерланга B')
plt.xlabel('Навантаження (Erlangs)')
plt.ylabel('Ймовірність втрати')
plt.legend()
plt.grid(True)
plt.show()
```
