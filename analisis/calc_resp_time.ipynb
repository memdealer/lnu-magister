```python
import matplotlib.pyplot as plt
import pandas as pd

# Завантаження даних з CSV
file_path = 'data/response_time_data.csv'  # Вкажіть шлях до вашого файлу
data = pd.read_csv(file_path)

# Зчитування даних
requests = data['Requests']
rest_response_time = data['REST_Response_Time']
grpc_response_time = data['gRPC_Response_Time']

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(requests, rest_response_time, label='REST API', color='red')
plt.plot(requests, grpc_response_time, label='gRPC', color='green')
plt.title('Залежність часу відповіді від кількості одночасних запитів')
plt.xlabel('Кількість одночасних запитів')
plt.ylabel('Час відповіді (мс)')
plt.legend()
plt.grid(True)
plt.show()
```
