```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження даних з CSV
file_path = 'data/response_time_data_sorted.csv'  # Замініть шлях, якщо файл знаходиться в іншому місці
data = pd.read_csv(file_path)


requests = data['Requests']
response_time_http1 = data['Response_Time_HTTP1']
response_time_http2 = data['Response_Time_HTTP2']

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(requests, response_time_http1, label='HTTP/1.1 без мультиплексування', color='red')
plt.plot(requests, response_time_http2, label='HTTP/2 з мультиплексуванням', color='green')
plt.xlabel('Кількість запитів')
plt.ylabel('Час відгуку (секунди)')
plt.title('Графік часу відгуку')
plt.legend()
plt.grid(True)
plt.show()
```
